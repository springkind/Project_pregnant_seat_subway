### 0. 전체 앉아있던 사람들 중 임산부 비율

```python
empty = pt[pt['성별'] == '비었음']
```

```python
not_empty = pt[pt['성별'] != '비었음']
```

```python
len(empty)
```

```python
len(not_empty)
```

```python
len(pt) == len(empty) + len(not_empty)
```

```python
pragnant = pt[(pt['성별'] != '비었음') & (pt['임산부 여부'] == 'O')]
```

```python
len(pragnant)
```

```python
f'임산부석 자리에 앉았던 {len(not_empty)}명 중 뱃지 유무로 또는 육안으로 구분 가능한 임산부는 {len(pragnant)}명이었습니다'
```

```python
f'자리가 비어있지 않은 경우 중 명확히 구분 가능한 임산부가 앉아있는 비율은 {round(len(pragnant)/len(not_empty)*100,1)}% 입니다.'
```

### 3. 각 호선별로 평균 몇 구간이 비어있었는지

```python
empty_times = pd.DataFrame(empty.groupby(['호선', '날짜'])['기록 시간'].count()).rename(columns={'기록 시간': '빈 횟수'})
```

```python
empty_times
```

```python
empty_tiems_sum = empty_times.groupby(['호선']).sum().reset_index()
```

```python
empty_tiems_sum
```

```python
empty_total = pd.DataFrame(pd.to_numeric(empty.groupby('호선')['비어있었던 역 수'].sum(), downcast='signed')).reset_index()
```

```python
empty_total
```

```python
empty_total['하루평균빈자리'] = empty_total['비어있었던 역 수']/empty_tiems_sum['빈 횟수']

# 이렇게 해도 됨.
# empty_set = pd.DataFrame(pd.merge(empty_total, empty_tiems_sum)) 
# empty_set['비어있던 구간의 갯수/하루'] = empty_set['비어있었던 역 수']/empty_set['빈 횟수']
```
